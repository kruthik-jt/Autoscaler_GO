```python
import os

# Get the path to the Autoscaling folder
autoscaling_folder_path = "/content/drive/My Drive/Autoscaling"

# Get a list of all the files in the Autoscaling folder
autoscaling_files = os.listdir(autoscaling_folder_path)

# Create a dictionary to store the files
autoscaling_files_dict = {}

# Read each file and store it in the dictionary
for filename in autoscaling_files:
  file_path = os.path.join(autoscaling_folder_path, filename)
  with open(file_path, "r") as f:
    file_contents = f.read()
    autoscaling_files_dict[filename] = file_contents
```

```python
def extract_numbers_from_even_lines(filename):
  """
  Extracts all the numbers from every even number line in the file and adds it to an array.

  Args:
    filename: The name of the file to read.

  Returns:
    An array of arrays, each containing the 4 numeric values separated by a comma present in every even number line.
  """

  # Get the file contents from the dictionary
  file_contents = autoscaling_files_dict[filename]

  # Split the file contents into lines
  lines = file_contents.split("\n")

  # Initialize an array to store the extracted numbers
  numbers_array = []

  # Iterate over the lines, starting from the second line (index 1)
  for i in range(1, len(lines), 2):
    # Split the line into words
    words = lines[i].split()

    # Extract the 4 numeric values from the line
    numbers = [float(word) for word in words if word.isdigit() or "." in word]

    # Add the numbers to the array
    numbers_array.append(numbers)

  # Return the array of numbers
  return numbers_array
```

```python
PIDRecord_E1_0_FIFA = extract_numbers_from_even_lines('PIDRecord_E1_0_FIFA.txt')
PIDRecord_E1_1_FIFA = extract_numbers_from_even_lines('PIDRecord_E1_1_FIFA.txt')
PIDRecord_E1_2_FIFA = extract_numbers_from_even_lines('PIDRecord_E1_2_FIFA.txt')

PIDRecord_E2_0_NASA = extract_numbers_from_even_lines('PIDRecord_E2_0_NASA.txt')
PIDRecord_E2_1_NASA = extract_numbers_from_even_lines('PIDRecord_E2_1_NASA.txt')
PIDRecord_E2_2_NASA = extract_numbers_from_even_lines('PIDRecord_E2_2_NASA.txt')

PIDRecord_E3_0_RandomFIFA = extract_numbers_from_even_lines('PIDRecord_E3_0_RandomFIFA.txt')
PIDRecord_E3_1_RandomFIFA = extract_numbers_from_even_lines('PIDRecord_E3_1_RandomFIFA.txt')
PIDRecord_E3_2_RandomFIFA = extract_numbers_from_even_lines('PIDRecord_E3_2_RandomFIFA.txt')

PIDRecord_E4_0_NASARepeat = extract_numbers_from_even_lines('PIDRecord_E4_0_NASARepeat.txt')
PIDRecord_E4_1_NASARepeat = extract_numbers_from_even_lines('PIDRecord_E4_1_NASARepeat.txt')
PIDRecord_E4_2_NASARepeat = extract_numbers_from_even_lines('PIDRecord_E4_2_NASARepeat.txt')

MARecord_E1_FIFA = extract_numbers_from_even_lines('MARecord_E1_FIFA.txt')
MARecord_E2_NASA = extract_numbers_from_even_lines('MARecord_E2_NASA.txt')
MARecord_E3_RandomFIFA = extract_numbers_from_even_lines('MARecord_E3_RandomFIFA.txt')
MARecord_E4_NASARepeat = extract_numbers_from_even_lines('MARecord_E4_NASARepeat.txt')
```

```python
import matplotlib.pyplot as plt
# Extract the first elements from the arrays
pid_record_e1_1_fifa = [element[0] for i, element in enumerate(PIDRecord_E1_1_FIFA) if i % 5 == 0]
ma_record_e1_fifa = [element[0] for element in MARecord_E1_FIFA]

# Create a time axis for the PIDRecord_E1_1_FIFA data
time_axis_pid = range(0, len(pid_record_e1_1_fifa))

# Create a time axis for the MARecord_E1_FIFA data
time_axis_ma = range(0, len(ma_record_e1_fifa))

# Plot the time series graph
plt.plot(time_axis_pid, pid_record_e1_1_fifa, color='blue', label='PID')
plt.plot(time_axis_ma, ma_record_e1_fifa, color='red', label='Moving Average')

# Add a legend and title
plt.legend()
plt.xlabel('Time (5=24mins)')
plt.ylabel('Underprovisioning Accuracy (%)')
plt.title('Time Series Graph of FIFA Experiment')

# Show the plot
plt.show()
```
